```python
import requests
import json
from bs4 import BeautifulSoup
from selenium import webdrivers
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
import re as re
import time
import pandas as pd
import os
```

```python
cars= pd.read_csv("cars.csv")
cars.drop(columns="Unnamed: 0",inplace=True)
b= cars["car_brand"].unique()
```

```python
driver = webdriver.Chrome(executable_path='chromedriver') #executable_path="path to your chromedriver"
# 0. Declaring the options
options = webdriver.ChromeOptions()
#options.add_argument('--headless') # It will prevent chrome from opening
options.add_argument('--incognito')
options.add_argument("--window-size=1260,800")
# 1. Initializing driver
driver = webdriver.Chrome(options = options) #executable_path="path to your chromedriver"
# 2. My link
url = "https://uae.yallamotor.com/used-cars/search?sort=updated_desc"
driver.get(url)  # requests.get(url)
time.sleep(2)
start = time.time()
last_height = driver.execute_script("return document.body.scrollHeight")
lis=[]
car_brand=[]
car_model=[]
car_price=[]
car_odo=[]
car_year=[]
carmb=[]
carodo=[]
carp=[]
while True:
    element = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[4]/div/div[1]/div[3]/div/a[7]')
    desired_y = (element.size['height'] / 2) + element.location['y']
    current_y = (driver.execute_script('return window.innerHeight') / 2) + driver.execute_script('return window.pageYOffset')
    scroll_y_by = desired_y - current_y
    driver.execute_script("window.scrollBy(0, arguments[0]);", scroll_y_by)
    time.sleep(5)
    html = driver.execute_script("return document.body.outerHTML;")
    soup = BeautifulSoup(html, "html.parser")
    carmb += soup.find_all("a", {"class":"black-link"})
    cary = soup.find_all("a", {"class":"black-link"})
    carodo = soup.find_all("div", {"class":"m6l"})
    carp = soup.find_all("span", {"class":"font24"})
    for i in carmb:
        c = i.text
        c=c.split(' ')
        c=c[1:]
    for bb in c:
        for a in b:
            if bb in a:
                lis.append(str(c.pop(0)))
    z=" ".join(lis)
    z=z.strip()
    lis=[]
    car_brand.append(z)
    for i in cary:
        x = i.text
        x=x.split(' ')  
        car_year.append(int(x.pop()))
    for i in carp:
        p= i.text
        p=p.replace(",",'')
        p.strip()
        car_price.append(int(p))
    for i in carodo:
        o= i.text
        o=o.replace(",",'')
        o.strip()
        car_odo.append(int(o))
    p = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[4]/div/div[1]/div[3]/div/a[7]')
    p.click()
    
time.sleep(1)
```
